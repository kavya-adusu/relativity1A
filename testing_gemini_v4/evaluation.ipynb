# Distribution of Candidate Scores


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data into a DataFrame
csv_file_path = 'aggregated_results.csv'  # Replace with your actual path
df = pd.read_csv(csv_file_path)

# Count the number of occurrences of each score
score_counts = df['score'].value_counts().sort_index()

# Plotting the bar chart
plt.figure(figsize=(8, 6))
plt.bar(score_counts.index, score_counts.values, color='skyblue')
plt.xlabel('Score')
plt.ylabel('Number of Candidates')
plt.title('Distribution of Candidate Scores')
plt.xticks([1, 2, 3, 4])  # Ensure x-ticks show the score values 1-4
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Show the plot
plt.show()
```

# Average Score by Candidate Group

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data into a DataFrame
csv_file_path = 'aggregated_results.csv'  # Replace with your actual path
df = pd.read_csv(csv_file_path)

# Extract the prefix from the 'set_name' column to create a 'group' column
df['group'] = df['set_name'].apply(lambda x: x.split('_')[0])

# Calculate the average score for each group
group_avg_scores = df.groupby('group')['score'].mean().sort_values()

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(group_avg_scores.index, group_avg_scores.values, color='pink')
plt.xlabel('Candidate Group')
plt.ylabel('Average Score')
plt.title('Average Score by Candidate Group')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data into a DataFrame
csv_file_path = 'aggregated_results.csv'  # Replace with your actual path
df = pd.read_csv(csv_file_path)

# Extract the prefix from the set_name to create the candidate group
df['group'] = df['set_name'].apply(lambda x: x.split('_')[0])

# Plotting the scatter plot
plt.figure(figsize=(14, 8))
plt.scatter(df['group'], df['score'], alpha=0.6, color='purple')
plt.xlabel('Candidate Group')
plt.ylabel('Score')
plt.title('Scatter Plot of Scores by Candidate Group')
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Display the plot
plt.show()
```
